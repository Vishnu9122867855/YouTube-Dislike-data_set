```python
import ipywidgets as widgets
```

```python
h=widgets.Dropdown(
    options=video["channel_title"].unique(),
    value="NBC",
    description='channel:',
    disabled=False,
)

def piegraph(ch):
    nbc=video[video["channel_title"]==ch]
    lk=nbc["likes"].sum()
    dk=nbc["dislikes"].sum()
    cc=nbc["comment_count"].sum()
    plt.pie(x=[lk,dk,cc],labels=["likes","dislikes","comments"],autopct="%1.1f%%")
    plt.show()
    
```

```python
widgets.interact(piegraph,ch=h)
```
